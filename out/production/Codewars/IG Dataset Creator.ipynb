```python
import json
import os
from pprint import pprint
import pandas as pd
```

```python
dir_path = r"inbox"
wordList = []
for subdir, dirs, files in os.walk(dir_path):
    
    for file in files:
        if file.endswith(".json"):
            f = open(r'{}\{}'.format(subdir, file))
            data = json.load(f)
            
            for key, val in data['messages'][0].items():
                if key == "content":
                    print("Content is {} : {}".format(key, val))
                    wordList.append(val)
            
#             print(json.dumps(,
#                              sort_keys=True, indent=4))
df = pd.DataFrame(wordList)
```
